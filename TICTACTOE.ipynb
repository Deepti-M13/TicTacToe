```python
def display():
    print("        TICTACTOE        ")
    print("-------------------------")
    print("|       |       |       |")
    print(f"|   {mlist[1]}   |   {mlist[2]}   |   {mlist[3]}   |")
    print("|       |       |       |")
    print("-------------------------")
    print("|       |       |       |")
    print(f"|   {mlist[4]}   |   {mlist[5]}   |   {mlist[6]}   |")
    print("|       |       |       |")
    print("-------------------------")
    print("|       |       |       |")
    print(f"|   {mlist[7]}   |   {mlist[8]}   |   {mlist[9]}   |")
    print("|       |       |       |")
    print("-------------------------")
def playermarkers(player):
    # use while loop until player chooses x or o and return markers
    #return tuple
    while True:
        ch=input("ENTER YOUR CHOICE 'X' OR 'O'").upper()
        if player==1 and ch=='X' or player==2 and ch=='O':
            return('X', 'O')
        elif player==2 and ch=='X'or player==1 and ch=='O':
            return('O', 'X')
            
def continue_playing():
    # use while until player choose yes or no and return boolean
    while True:
        ch=input("DO YOU WANT TO CONTINUE PLAYING 'YES' OR 'NO'").upper()
        if(len(ch)>=1):
            if ch[0]=='Y':
                return True
            elif ch[0]=='N':
                return False
                
def choosepositionandmark(pm):
    #use while until player chooses a position from 1 to 9
    #if postion in (1,9) and position available mark position
    while True:
        pos=(input("ENTER YOUR POSITION FROM 1 TO 9"))
        if pos.isdigit():
            pos=int(pos)
            if  pos in range(1,10) and mlist[pos]==" ":
                mlist[pos]=pm
                break
                
def gameover(marker):
    #returns boolean
    if (mlist[1]==mlist[2]==mlist[3]==marker or mlist[4]==mlist[5]==mlist[6]==marker
    or mlist[7]==mlist[8]==mlist[9]==marker or mlist[1]==mlist[4]==mlist[7]==marker
    or mlist[2]==mlist[5]==mlist[8]==marker or mlist[3]==mlist[6]==mlist[9]==marker
    or mlist[1]==mlist[5]==mlist[9]==marker or mlist[3]==mlist[5]==mlist[7]==marker):
        return True

def boardfull(mlist):
    for i in range(1,10):
        if mlist[i]==" ":
            return False
        else:
            pass
    return True
        
def playagain():
    #returns boolean
    while True:
        ch=input("DO YOU WANT TO PLAY AGAIN 'YES' OR 'NO'").upper()
        if(len(ch)>=1):
            if ch[0]=='Y':
                return True
            elif ch[0]=='N':
                return False
    
```

```python
from random import randint
from IPython.display import clear_output
while True:
    clear_output()
    mlist=[" "]*10
    player=randint(1,2)
    print(f"PLAYER {player} WILL PLAY FIRST")
    p1m, p2m=playermarkers(player)   
    gameon=continue_playing()
    if gameon==False:
        break    
    while gameon:
        if player==1:
            clear_output()
            display()
            choosepositionandmark(p1m)
            display()
            if gameover(p1m):
                #display congrats player 1 you won the game
                clear_output()
                display()
                print("PLAYER 1 WON THE GAME")
                break
            elif boardfull(mlist):
                #display game draw
                clear_output()
                display()
                print("THE GAME IS A DRAW")
                break
            else:
                player=2
        if player==2:
            clear_output()
            display()
            choosepositionandmark(p2m)
            display()
            if gameover(p2m):
                #display congrats player 2 you won the game
                clear_output()
                display()
                print("PLAYER 2 WON THE GAME")
                break
            elif boardfull(mlist):
                #display game draw
                clear_output()
                display()
                print("THE GAME IS A DRAW")
                break
            else:
                player=1
    again=playagain()
    if again==False:
        break
            
        
    
```
